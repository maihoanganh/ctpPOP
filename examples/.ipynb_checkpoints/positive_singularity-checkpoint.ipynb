```julia
using DynamicPolynomials

println("***Problem setting***")

n=20

r=n
#r=ceil(Int64,n/4)

println("Number of variable: n=",n+r)
println("====================")

@polyvar v[1:r] x[1:r]# variables


# random quadratic objective function f

#=M=[-0.3 -0.2 -0.2 -0.4 0.5 -0.1;
0.8 0.1 0 0.2 0 -0.2;
-0.4 -0.4 0.2 -0.1 0.2 -0.5;
0.1 0.7 0.7 -0.2 -0.1 -0.1;
-0.4 0 -0.5 0 -0.3 -0.5;
0 0.5 0 0.6 0.1 0.1]=#

M=2*rand(Float64,n,r).-1

f=-sum((M*v).^2)


# unit sphere constraint
g=Vector{Polynomial{true,Float64}}(v) #type of coefficients of each polynomial must be float
#g=Vector{Polynomial{true, Float64}}([])
m=length(g)
println("Number of inequality constraints: m=",m)
println("====================")


h=[1.0-sum(v.^2)]

l=length(h)
println("Number of equality constraints: l=",l)
println("====================")

k=2

println("Relaxed order: k=",k)
```

```julia
include("../src/SpectralPOP.jl")
using .SpectralPOP

opt_val=SpectralPOP.SumofSquares_POP(v,f,g,h,k)
```
