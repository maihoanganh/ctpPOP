```julia
using LinearAlgebra

n=10

r=10*n

A=2*rand(Float64,r,r).-1

B=2*rand(Float64,r,r).-1

C=2*rand(Float64,r,r).-1

D=2*rand(Float64,r,r).-1

#=A=2*A
B=2*B
C=2*C
D=2*D=#

M=zeros(Float64,n*r,n*r)

for i in 1:n, j in 1:i
    if i==j
        M[r*(i-1)+1:i*r,r*(j-1)+1:j*r]=D
    else
        M[r*(i-1)+1:i*r,r*(j-1)+1:j*r]=C*A^(i-j-1)*B
    end
end

#M./sqrt(norm(M'*M))
```
